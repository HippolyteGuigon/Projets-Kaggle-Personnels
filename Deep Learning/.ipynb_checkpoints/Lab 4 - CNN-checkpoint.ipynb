# 2 - First naive model

In order to better understand the importance of CNNs, it is instructive to first see how well a naive dense network performs on the dataset.

**7) Create a sequential model with 4 `Dense` hidden layers of 2048, 1024, 512, and 256 nodes each, with ReLU activation, and a linear output layer of 10 nodes. Compile the model with a `categorical_crossentropy` loss, using the SGD optimizer, including the `accuracy` metric. 
Note that you will need to use the `Flatten` layer first in order to convert the 3D (x, y, rgb) image data into 1D.**

```python
model = tf.keras.Sequential()
model.add(Flatten(input_shape=(3, 32, 32)))
model.add(tf.keras.layers.Dense(2048, activation='relu'))
model.add(tf.keras.layers.Dense(1024, activation='relu'))
model.add(tf.keras.layers.Dense(512, activation='relu'))
model.add(tf.keras.layers.Dense(256, activation='relu'))
model.add(tf.keras.layers.Dense(10, activation='relu'))

print(model.output_shape)

model.compile(optimizer='SGD', loss='categorical_crossentropy',metrics=['accuracy'])
```

**8) Compute by hand the total number of trainable parameters (weights and biases) in the model.**

$(3072+1) * 2048 + (2048+1) * 1024 + (1024 +1) * 512 + (512+1) * 256 + (256+1)* 10 = 9,050,378$ 

There are $9,050,378$ parameters

## 3.1 - Create your first CNN

**12) Create a new model with the following layers (use the same optimizer and loss as above)**
  - 3x3 2D convolution with zero padding (same), 32 filters
  - ReLU activation
  - 3,3 2D convolution, no padding, 32 filters
  - ReLU activation
  - Max pooling with size (2,2)
  - 3x3 2D convolution, no padding, 64 filters
  - ReLU activation
  - 3x3 2D convolution, no padding, 64 filters
  - ReLU activation
  - Max pooling with size (2,2)
  - Flatten
  - Dense layer with 512 nodes, ReLU activation
  - Softmax output layer with 10 nodes

```python
model_CNN = tf.keras.Sequential()
model_CNN.add(tf.keras.layers.Conv2D(32, (3, 3), padding="same", activation="relu"))
model_CNN.add(tf.keras.layers.Conv2D(32, (3, 3), activation="relu"))
model_CNN.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
model_CNN.add(tf.keras.layers.Conv2D(64, (3, 3), activation="relu"))
model_CNN.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
model_CNN.add(tf.keras.layers.Conv2D(64, (3, 3), activation="relu"))
model_CNN.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
model_CNN.add(tf.keras.layers.Flatten())
model_CNN.add(tf.keras.layers.Dense(512, activation="relu"))
model_CNN.add(tf.keras.layers.Dense(10, activation='softmax'))
model_CNN.compile(optimizer='adam', loss=tf.keras.losses.categorical_crossentropy,metrics=['accuracy'])
```

**13) Compute by hand the number of trainable parameters in this network.  Are there more or less than the more simple dense network?  Why?  Confirm with `summary()`.**

For conv2d Layers, we use this formula : $(m*n*d + 1)*k $ where ($m$ width, $n$ height, $k$ number of filters of the current layer and $d$ number of filters in the previous layer) 

We will compute the number of parameters on each layer : 

$(3*3*1)+1)*32  = 320 $

$((3*3*32)+1)*32= 9248$

$((3*3*32)+1)*64 = 18496$

$((3*3*64)+1)*64 = 36928$

$(256+1) * 512 = 131584 $ 

$ (512+1)*10 = 5136$

Then we sum all of the results : there are 201,706 parameters.

There are less parameters due to sparsity (all networks are not connected to each other).

**14) Train the network for 10 epochs, with batch size of 32. How does the validation accuracy change with each epoch?**

```python
model_CNN.fit(X_train, y_train, batch_size=32, epochs=10, validation_data=(X_test, y_test))
```

```python
scores_CNN = model_CNN.evaluate(X_test, y_test)
```

```python
print("With this CNN model, loss is {:.2f} and accuracy is {:.2f}%".format(scores_CNN[0], 100*scores_CNN[1]))
```

The valiation accuracy increases with each epoch

**16) Note how the validation accuracy begins to decrease at some point, while the training accuracy continues to increase.  What is this phenomenon?  Try adding 3 dropout layers to the model, one before each max pooling layer and one before the last layer, using a dropout ratio of 0.25.  Does this improve the model? How does the performance vary with the dropout ratio?**

This phenomenon is called "overfitting" and can be explained by the fact that having excessive information makes it difficult to generalize

```python
model_dropout = tf.keras.Sequential()
model_dropout.add(tf.keras.layers.Conv2D(32, (3, 3), padding="same", activation="relu"))
model_dropout.add(tf.keras.layers.Conv2D(32, (3, 3), activation="relu"))
model_dropout.add(tf.keras.layers.Dropout(.25))
model_dropout.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
model_dropout.add(tf.keras.layers.Conv2D(64, (3, 3), activation="relu"))
model_dropout.add(tf.keras.layers.Dropout(.25))
model_dropout.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
model_dropout.add(tf.keras.layers.Conv2D(64, (3, 3), activation="relu"))
model_dropout.add(tf.keras.layers.Dropout(.25))
model_dropout.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
model_dropout.add(tf.keras.layers.Flatten())
model_dropout.add(tf.keras.layers.Dense(512, activation="relu"))
model_dropout.add(tf.keras.layers.Dropout(.25))
model_dropout.add(tf.keras.layers.Dense(10, activation='softmax'))
model_dropout.compile(optimizer='SGD', loss='categorical_crossentropy',metrics=['accuracy'])

model_dropout.fit(X_train, y_train, batch_size=32, epochs=10, validation_data=(X_test, y_test))

scores_dropout = model_dropout.evaluate(X_test, y_test)

print("The loss with 0.25 dropout is {:.2f} and the accuracy is {:.2f}%".format(scores_dropout[0], 100*scores_dropout[1]))
```

```python
Dropout = [0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6]
scores_graph = []

for i in Dropout:
    model_dropout = tf.keras.Sequential()
    model_dropout.add(tf.keras.layers.Conv2D(32, (3, 3), padding="same", activation="relu"))
    model_dropout.add(tf.keras.layers.Conv2D(32, (3, 3), activation="relu"))
    model_dropout.add(tf.keras.layers.Dropout(i))
    model_dropout.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
    model_dropout.add(tf.keras.layers.Conv2D(64, (3, 3), activation="relu"))
    model_dropout.add(tf.keras.layers.Dropout(i))
    model_dropout.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
    model_dropout.add(tf.keras.layers.Conv2D(64, (3, 3), activation="relu"))
    model_dropout.add(tf.keras.layers.Dropout(i))
    model_dropout.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
    model_dropout.add(tf.keras.layers.Flatten())
    model_dropout.add(tf.keras.layers.Dense(512, activation="relu"))
    model_dropout.add(tf.keras.layers.Dropout(i))
    model_dropout.add(tf.keras.layers.Dense(10, activation='softmax'))
    model_dropout.compile(optimizer='SGD', loss='categorical_crossentropy',metrics=['accuracy'])

    model_dropout.fit(X_train, y_train, batch_size=32, epochs=10, validation_data=(X_test, y_test))

    scores = model_dropout.evaluate(X_test, y_test)
    print("\n")
    print(100*scores[1])
    scores_graph.append(100*scores[1])
```

```python
plt.figure(figsize=(10, 6))
plt.title("Average score per Dropout rate")
plt.plot([0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6], scores_graph, marker='o', linestyle='-' )
plt.xticks([0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6])
plt.xlabel("Dropout rate")
plt.ylabel("Accuracy per dropout rate (%)");
```

Thus, even if it is not an absolute correlation, accuracy tends to get lower as dropout rate is higher

# 4 - Pretrained Networks

Note that several pre-trained networks are directly accessible via keras (see https://nbviewer.jupyter.org/github/fchollet/deep-learning-with-python-notebooks/blob/master/5.3-using-a-pretrained-convnet.ipynb)

**20) Build a classifier with a better accuracy on the test set than all CNN you have built before. One rule only: do not use CNN pretrained on CIFAR10.** 

```python
from keras.applications import ResNet50V2
from keras.layers import UpSampling2D


model_Resnet50 = tf.keras.Sequential()
conv_base = ResNet50V2(weights='imagenet',
                  include_top=False)
conv_base.trainable = False
model_Resnet50.add(UpSampling2D(size=(3,3)))
model_Resnet50.add(conv_base)
model_Resnet50.add(Flatten())
model_Resnet50.add(Dense(512, activation='relu'))
model_Resnet50.add(Dense(256, activation='relu'))
model_Resnet50.add(Dense(10, activation="softmax"))
model_Resnet50.compile(optimizer='adam', loss='categorical_crossentropy',metrics=['accuracy'])
early_stop = EarlyStopping(monitor="loss", patience=2, verbose=0, mode="min")
model_Resnet50.fit(X_train, y_train, batch_size=32, epochs=10, validation_split=0.2, callbacks=[early_stop])

scores_Resnet50 = model_Resnet50.evaluate(X_test, y_test)
```

```python
print("With this Resnet50V2 model, loss is {:.2f} and accuracy is {:.2f}%".format(scores_Resnet50[0], 100*scores_Resnet50[1]))
```
