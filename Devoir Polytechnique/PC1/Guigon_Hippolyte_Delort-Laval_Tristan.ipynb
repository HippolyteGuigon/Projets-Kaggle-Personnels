```python
#Exercise 4

def ceasar_code(your_string):
    alphabet = string.ascii_lowercase
    your_string = your_string.lower()
    mod_str = ""
    for i in your_string:
        try:
            mod_str += alphabet[(alphabet.index(i)+3)%26]
        except ValueError:
            mod_str += i
    return mod_str
                                
ceasar_code("a by d, b by e, z by c")
```

```python
#Exercise 5

import numpy as np

def multiplier(n):
    produit = 2
    for i in range(1,n+1):
        produit *= (4 * i ** 2)/(4 * i ** 2 - 1)
    return produit

def multiplicateur_numpy(n):
    a = np.arange(1,n+1,1)
    multiplicateur = (4 * a ** 2)/(((4 * a ** 2) -1))
    return 2 * np.prod(multiplicateur) 

print("The product value without Numpy method is: {:.7f}".format(round(multiplier(10000),7)))
print("\n")
print("The product value with the Numpy method is: {:.7f}".format(round(multiplicateur_numpy(10000),7)))
        
```

```python
#Exercise 7

from numpy.linalg import eig 

M = np.random.uniform(-1,1,size=(5,20))
G = np.dot(M.T,M)

print(M)

D , V = eig(G)

print("\n")
print("The Euclidian norm of G is {:.2f}".format(np.linalg.norm(G,2)))
print("\n")
print("The L1 norm of G is {:.2f}".format(np.linalg.norm(G,1)))
print("\n")

if np.all(G.T == G) and np.any(D<0):
    print("The matrix G is symetric, but not positive definite")
elif np.any(G.T != G):
    print("The matrix G is not symetric")
elif np.all (G.T == G) and np.all(D>0):
    print("The matrix G is symetric positive definite")
else: 
    print("The matrix G is symetric positive semi-definite")
```

```python
#Exercise 9

from os import path
import pandas as pd
import urllib
import zipfile
import sys

url = u'https://archive.ics.uci.edu/ml/machine-learning-databases/00235/'
filename ='household_power_consumption'
zipfilename = filename +'.zip'
Location = url + zipfilename

if not(path.isfile('zipfilename')):
    urllib.request.urlretrieve(Location, zipfilename)

zip = zipfile.ZipFile(zipfilename)
zip.extractall()
na_values = ['?','']
fields = ['Date','Time','Global_active_power']
df = pd.read_csv(filename +'.txt', sep=';', nrows=200000,na_values=na_values, usecols=fields)
```

```python
#Exercise 10

print("The number of missing values is {}".format(df["Global_active_power"].isna().sum()))
```

```python
#Exercise 11

df = df.dropna()
df.head()
```

```python
#Exercise 12

from datetime import datetime

df = df.set_index(pd.to_datetime(df["Date"] + " " + df["Time"]))
df["Cleaned_Date"] = pd.to_datetime(df["Date"] + " " + df["Time"])
df.drop(["Date","Time","Cleaned_Date"],axis = 1)
```

```python
#Exercise 13

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()

df["Date_2"] = pd.to_datetime(df["Date"])
power_daily_averages = df.groupby('Date_2').mean()

plt.figure(figsize=(20,10))
plt.plot(power_daily_averages)
plt.title("Daily Power Averages in 2007", fontsize=20)
start = '2007-01-01'
end = '2007-04-30'
start_dt = datetime.strptime(start, '%Y-%m-%d')
end_dt = datetime.strptime(end, '%Y-%m-%d')
plt.xlim(start_dt, end_dt)
plt.xlabel('Date', fontsize=20)
plt.ylabel('Power Consumption', fontsize=20);

#Electricity consumption is inversely proportional to the temperature.
#When temperatures fall at the beginning of February, average power consumption spikes up.
#At the very beginning of march, power consumption rises because of a cold snap probably.
#Conversely, at the very end of March, temperatures spike up whereas average power consumption falls.
```
